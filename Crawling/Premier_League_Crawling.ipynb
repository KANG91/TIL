```python
import pandas as pd
```

```python
from  selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.premierleague.com/tables")
```

```python
df = pd.DataFrame(columns = ["Position", "Club", "Played", "Won", "Drawn", "Lost", "GF", "GA", "GD", "Point"])
```

```python
# driver.find_element_by_css_selector("#advertClose").click()
```

```python
tbody = driver.find_elements_by_css_selector("tbody.tableBodyContainer tr:not(expandable)")[ : 39]
tbody = [each for idx, each in enumerate(tbody) if idx % 2 == 0 ]
```

```python
df
```

```python
for each in tbody : 
    position = each.find_elements_by_css_selector("span.value")[0].text
    club_name = each.find_elements_by_css_selector("a span.long")[0].text
    played = each.find_elements_by_css_selector("td:nth-child(4)")[0].text
    won = each.find_elements_by_css_selector("td:nth-child(5)")[0].text
    drawn = each.find_elements_by_css_selector("td:nth-child(6)")[0].text
    lost = each.find_elements_by_css_selector('td:nth-child(7)')[0].text
    GF = each.find_elements_by_css_selector("td:nth-child(8)")[0].text
    GA = each.find_elements_by_css_selector("td:nth-child(9)")[0].text
    GD = each.find_elements_by_css_selector("td:nth-child(10)")[0].text
    point = each.find_elements_by_css_selector("td:nth-child(11)")[0].text
    
    data = {
        "Position" : position,
        "Club" : club_name,
        "Played" : played,
        "Won" : won,
        "Drawn" : drawn,
        "Lost" : lost,
        "GF" : GF,
        "GA" : GA,
        "GD" : GD,
        "Point" : point,
     }
    df.loc[len(df)] = data

driver.close()
```
