### 인터넷을 이해하는 핵심 key : "IP주소"
Internet Protocol의 약자. 각각의 플레이어들이 준수해야 하는 규칙들을 '인터넷 프로토콜', 줄여서 IP라고 함. 인터넷을 이용하기 위해 연결되어 있는 여러 컴퓨터들의 주소를 <strong>IP주소</strong>라고 함.  
서로 연락하기 위해 각자 전화번호가 있듯 다른 컴퓨터에 컨택하기 위해서 IP주소가 있음.

웹사이트에 접속하는 방법은 2가지
- 1. 아이피 주소
- 2. 도메인 주소(ex.www.google.com)

실제로 통신이 일어나기 위해선 서버쪽 컴퓨터, 서버에 접속하는 클라이언트 모두 아이피주소를 가지고 있어야함.(전화를 할 때, 전화를 하는 쪽 받는쪽 모두 전화번호가 있어야 하듯)  

자신의 ip 확인하는 방법 - 검색엔진에서 'my ip' 검색. 

클라이언트가 서버에게 요청 후 응답을 받음.(모바일폰, 노트북, 데스크탑 모두 같은 원리로 움직임)  

편지로 비유를 하자면, 보내는 사람 : 받는 사람 = 클라이언트 아이피 : 서버 아이피
"서버 아이피의 홈페이지를 보고 싶습니다"라는 내용과 같음.
서버가 요청을 받으면 이후, 다시 서버에서 클라이언트로 응답을 보냄(보내는 사람과 받는사람이 바뀌는 것 같은)  

Ip주소의 단점 - "기억하기 어렵다"  -> 도메인 주소를 이용한 접속

사실 컴퓨터는 도메인을 통해서 접속이 아닌 ip를 통해서'만' 접속할 수 있음.

DNS(Domain Name System) : IP의 이름이라 할 수 있는 도메인 이름이 작용하는 체계.  

54.64.81.78의 복잡한 아이피 주소로 접근하기 힘드니, 도메인인 "opentutorial'라는 도메인을 가지고 있음. 클라이언트는 도메인 "https://opentutorials/course"라는 주소를 타이핑 후 입력 -> 도메인 네임서버(Name Server) -> 우리의 컴퓨터는 도메인 네임서버를 알고 있으므로 open tutorials에 접속하기 이전에 먼저 도메인 네임 서버에 요청, opentutorial의 ip를 받고 다시 opentutorial로 요청 - 마침내 클라이언트가 html 언어로써 응답받음  

도메인을 사용하고 싶다면? - 도메인을 구입해서 네임서버에 등록!  




### 사설 ip, 공인 ip
ip의 체계 -> "0.0.0.0 ~ 255.255.255.255" 사이의 값들 -> 총 42억개의 경우의 수.(컴퓨터, 스마트폰, 사물인터넷 등의 아이피 포함) -> 아이피의 부족함.

서버에 접속할 수 없는 이유 -> 아이피 주소의 부족?

방법 1. IPv6 체계
IPv4(version 4) : 222.109.62.43
IPv6(version 6) : ........2의 128승개(2823구 6692양 0938자 4636해 6337경....) -> 새로운 주소 체계

기존의 체계를 사용하고 있었기 때문에, ipv4와 IPv6가 같이 사용되고 있음.

공유기란? -> 집에 컴터가 3대일 경우, 정액제인 통신망을 공유기에 연결 - 3대에 모두 통신망을 연결하지 않고도 인터넷 사용 가능 - 가격 절감

**중요한건 공유기도 아이피를 갖게 된다.**

마이 아이피를 통해 확인할 수 있는 아이피 주소는 **컴퓨터의 아이피가 아닌, 공유기의 아이피임**

마치 회사에 내선번호가 있어, 회사에 전화를 할 경우, 회사의 중개자가 전화를 먼저 받고 내선번호를 이용해 관게자와 연결할 수 있는 것과 유사.

중개자의 전화번호(공유기의 주소) : 공인 IP
내선번호(개별 컴퓨터의 아이피) : 사설 IP

**하나의 공인 아이피로 여러개의 컴퓨터가 인터넷을 사용할 수 있게하는 것 - 사설 IP**

공인 아이피기에 공유기에 할당 되어 있다면, 개별 컴퓨터로 접속이 불가.

포트포워딩 : 공유기의 아이피를 통해 접속한 후, 개별컴퓨터(웹서버가 설치되어 있는, 사설 아이피)로 토스해주는 것.

** 그럼 포트(Port)란 무엇인가?**

클라이언트가 특정 서버에 접속했을 때 '정박'하는 것. 
정해진 포트번호(ex.80)가 아닌 다른 포트번호(ex.1212)를 입력하면 서버에 접속 불가.

포트 포워딩(port forwarding) : 공인 아이피로 연결됬던 주소를 내가 원하는 특정, 개별 컴퓨터로 'toss' 할 수 있게 해주는 기능. 


### 포트포워딩
WAN - 공인 아이피를 부여 받음.
LAN1, 2, 3 - 각각의 컴퓨터가 연결되며, 서로 다른 아이피 주소를 부여 받음.

'192.168.0.1'을 통해 공유기 관리자에 접속할 수 있음. 

공유기에 연결된 컴퓨터의 사설 아이피를 알아야 함. 
외부 포트 - 공유기에 연결된 통신사에서 들어오는 케이블
내부 포트 - 공유기와 서버를 구동하려는 컴퓨터의 포트가 무엇인지. 

외부포트 - 공유기 - 내부포트를 만들어 주는 것

### 유동 IP와 DDNS
유동 IP :  
컴퓨터를 오랫동안 사용하지 않으면, 컴퓨터에게 부여했던 아이피를 통신사에서 회수. -> 새로운 가입자에게 제공  

컴퓨터를 사용하려면 '새로운 아이피' 부여  

사용자들이 예전 아이피로 접속하려고 하면, 다시 제공 - 한정된 아이피를 운영하는 방법  

DDNS(Dynamic DNS) :
유동 IP가 되어 예전 서버에 접속할 수 없을 때.
자신의 컴퓨터 아이피가 네임 서버에 내 아이피가 변했다고 넘겨 주는 것. - 네임 서버는 변한 내 아이피를 계속 보유 - 서버 접속에 문제 없음.

사용자는 도메인 네임으로 접속 - DDNS는 도메인과 호스트의 아이피를 지속적으로 동기화 - 사용자는 도메인을 통해 올바른 아이피로 접속할 수 있음
